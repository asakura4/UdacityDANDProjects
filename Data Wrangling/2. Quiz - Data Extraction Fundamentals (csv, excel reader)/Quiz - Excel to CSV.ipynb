```python
# -*- coding: utf-8 -*-
'''
Find the time and value of max load for each of the regions
COAST, EAST, FAR_WEST, NORTH, NORTH_C, SOUTHERN, SOUTH_C, WEST
and write the result out in a csv file, using pipe character | as the delimiter.

An example output can be seen in the "example.csv" file.
'''

import xlrd
import os
import csv
from zipfile import ZipFile

datafile = "2013_ERCOT_Hourly_Load_Data.xls"
outfile = "2013_Max_Loads.csv"


def open_zip(datafile):
    with ZipFile('{0}.zip'.format(datafile), 'r') as myzip:
        myzip.extractall()


def parse_file(datafile):
    workbook = xlrd.open_workbook(datafile)
    sheet = workbook.sheet_by_index(0)
    data = {}
    # YOUR CODE HERE
    sheet_data = [[sheet.cell_value(r, col) for col in range(sheet.ncols)] for r in range(sheet.nrows)]
    '''
    sheet_data = []
    for r in range(sheet.nrows):
        for col in range(sheet.ncols):
            sheet_data[r][col] = sheet.cell_value(r,col)
    '''
    data = []
    station = []
    maxRowNum = 0
    (year, month, day, hour, minute, second) = (0,0,0,0,0,0)
    station = [sheet_data[0][i] for i in range(1,sheet.ncols)]

    
    for i in range(1, sheet.ncols):
        newDict = {}
        zoneData = sheet.col_values(i, start_rowx = 1, end_rowx = sheet.nrows)
        max_load = max(zoneData)
        maxRowNum = zoneData.index(max(zoneData)) + 1
        (year, month, day, hour, minute, second) = xlrd.xldate_as_tuple(sheet.cell_value(maxRowNum, 0),0)
        newDict['Year'] = year
        newDict['Month'] = month
        newDict['Day'] = day
        newDict['Hour'] = hour
        newDict['Max Load'] = max_load
        newDict['Station'] = station[i-1]
        data.append(newDict)
    
    # Remember that you can use xlrd.xldate_as_tuple(sometime, 0) to convert
    # Excel date to Python tuple of (year, month, day, hour, minute, second)
    return data

def save_file(data, filename):
    with open(filename, 'w') as csvfile:
        fields = ['Station','Year', 'Month', 'Day', 'Hour', 'Max Load']
        csvWriter = csv.DictWriter(csvfile, delimiter = '|', fieldnames = fields)
        csvWriter.writeheader()
        for line in data:
            csvWriter.writerow(line)
            

    
if __name__ == "__main__":
    #open_zip(datafile)
    data = parse_file(datafile)
    save_file(data, outfile)

    number_of_rows = 0
    stations = []
```
